```python
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis, color="green", marker="D", linewidth=2, label="Boston")
# Create a legend
ax.legend()
#Create a title
ax.set_title("PyBer Fare by Month")
#Create label for x-axis
ax.set_xlabel("Date")
#Create y-axis label
ax.set_ylabel("Fare ($)")
#Set the y-axis limit
ax.set_ylim(0, 45)
#Add grid lines
ax.grid()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()
#Create a title
plt.title("PyBer Fare by Month")
#Create axes labels
plt.xlabel('Date')
plt.ylabel('Fare ($)')
#Add a legend
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
#invert the vertical axis
ax.invert_yaxis()
#Add a title
ax.set_title("PyBer Fare by Month")
#Add axes labels
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
```

```python
#Create a scatter plot
plt.plot(x_axis, y_axis, 'o')
```

```python
#Create a scatter plot
plt.scatter(x_axis, y_axis)
```

```python
plt.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color="skyblue", alpha=.8, s=[i * 5 for i in y_axis], linewidth=2, edgecolors="k", label="Boston")
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.invert_yaxis()
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
ax.set_xlim(0, 50)
```

```python
#Create a pie chart
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
#Pie Chart using Object-Oriented Interface
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis,
       labels=x_axis, 
       autopct='%.1f%%', 
       explode=explode_values, 
       shadow=True, 
       startangle=90, 
       counterclock=False, 
       colors=colors)
explode_values=(0, 0, 0.3, 0, 0, 0, 0.3, 0, 0, 0, 0, 0)
plt.show()
```
